## Addressing Missing Data/ Default Values

### As you cleaned your data, you may have observed missing data/ default values. Decide what to do with these values and explain your decision-making process.

Here are some methods to deal with missing values:
- Drop records with missing values
    - Check for biases induced by missing values - there may be trends in why something is missing
    - This requires domain knowledge to check for induced bias
- Imputation: Inferring missing values
    - Replace with average value (closest related subgroup mean)
    - Replace with a random value from the closest subgroup(Hot deck imputation) 
- Numerical: 
    - mean, mode, median
- Categorical
    - Mode, new category
- Directly model missing values during future analysis

### Now let's handle any outliers. Same process as before!

Here are some methods to deal with handling outliers:
- Drop records with outliers
    - Check for biases induced by missing values - there may be trends in why something is missing
    - This requires domain knowledge to check for induced bias
- Replace, as with Inferring missing values
    - Typically with min or max
- Binning
    - E.g. greater than 5

```python
#Visualization 1: Scatter plot

appl_df = df.where(df["Name"] == "Apple")
sns.set_style('whitegrid')
appl_g = sns.relplot(
    data = appl_df,
    x = "Date",
    y = "Low"
)

# sns.pairplot(appl_df)

plt.title("Apple Min/Max Price during Pandemic")
plt.show(appl_g)

#Your observations
# Graph is essentially what any finance app/stocks app would show if you selected the given time frame for the APPL stock ticker.
# With thousands of data points, the scatter plot essentially becomes a line graph, but can show some gaps which may be quicker
# jumps in price. 
```

```python
#Visualization 2 = horizontal bar plot
sns.set_theme(style="whitegrid")

f, ax = plt.subplots(figsize=(6, 15))

high_price = df['High']
low_price = df['Low']

sns.set_color_codes("pastel")

# high price
sns.barplot(
    x = "High",
    y = "Name",
    data = df,
    label = "High",
    color = 'g'
)

# low price
sns.barplot(
    x = "Low",
    y = "Name",
    data = df,
    label = "Low",
    color = 'r'
)

# legend, axes
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(ylabel = "Price (High & Low)", xlabel = "Company Name")
sns.despine(left=True, bottom=True)

#Your observations
# This visualization for low and high price of a stock is interesting but doesn't seem to say a lot. That being said,
# one thing I think that can be interpreted from this is that for Facebook and Apple, there was no overall increase in the
# timeframe that this dataset is made from, which is why there is no green bar that constitutes the high price values.
```

```python
#Visualization 3 Line Graph

sns.set_theme(style="darkgrid")

# Plot the responses for different events and regions
sns.lineplot(x="Date", y="Low",
             data=appl_df)

#Your observations
# This line graph shows the price of Apple stock from January 2020 to April 2022, accurately depicting the daily low price over time.
```
